```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#testrun: earth orbiting the sun
#aphelion 1 AU, eccentircy:  0.016710219, num of periods: 1,
#stepsize = 0.001, adaptive accuracy: 1e-5

#loadin and extracting data
fixed_data = np.loadtxt('Test/fixed.data')
adaptive_data = np.loadtxt('Test/adaptive.data')

time_fixed, x_fixed, y_fixed, _, _ = fixed_data.T
time_adaptive, x_adaptive, y_adaptive, _, _ = adaptive_data.T
x1 = [0,0]
y1 = [0,0]


fig, axs = plt.subplots(1, 2, figsize=(14, 6))


axs[0].plot(x_fixed, y_fixed, linestyle='--', label = "Fld plyja")
axs[0].scatter(x1,y1,color = "orange", s = 150, label = "Nap")
axs[0].set_title('RK4')
axs[0].set_xlabel('kitrs (x)')
axs[0].set_ylabel('kitrs (y)')
axs[0].grid(True)
axs[0].legend()
axs[0].axis('equal')


axs[1].plot(x_adaptive, y_adaptive, label = "Fld plyja")
axs[1].set_title('Adaptv RK4')
axs[1].scatter(x1,y1,color = "orange", s = 150, label = "Nap")
axs[1].set_xlabel('kitrs (x)')
axs[1].set_ylabel('kitrs (y)')
axs[1].grid(True)
axs[1].legend()
axs[1].axis('equal')

plt.tight_layout() 
plt.show()
```
