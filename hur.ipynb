```python
import numpy as np
import matplotlib.pyplot as plt

# Függvény az analitikus megoldás alapján
def y_analytical(x, c, L,t):
    return np.sin(np.pi * x / L) * np.cos(np.pi * c * t / L)

# Paraméterek
L = 1.0  # Húr hossza
c = 0.5  # Hullámsebesség

# x értékek
x_values = np.linspace(0, L, 1000)

# y értékek az analitikus megoldás alapján
y_values = y_analytical(x_values, c, L, 0)

# Ábrázolás
plt.plot(x_values, y_values, label="t = 0")
plt.xlabel("x")
plt.ylabel("y(x, 0)")
plt.title("1 módusú kezdőfeltétel analitikus megoldása t = 0-nál")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# y értékek az analitikus megoldás alapján
y_values = y_analytical(x_values, c, L, 2)

# Ábrázolás
plt.plot(x_values, y_values, label="t = 2")
plt.xlabel("x")
plt.ylabel("y(x, 2)")
plt.title("1 módusú kezdőfeltétel analitikus megoldása t = 2-nál")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

data = data[:, 0:100]

# Függvény az analitikus megoldás alapján
def y_analytical(x, c, L, t):
    return np.sin(np.pi * x / L) * np.cos(np.pi * c * t / L)

# Paraméterek
L = 1.0  # Húr hossza
c = 1.0  # Hullámsebesség

# x értékek
x_values = np.linspace(0, L, 100)

# Időpontok
t_values = np.linspace(0, 1, 100)

# Diszkrét adatok animációja
fig, ax = plt.subplots()

# Pontok száma
num_steps = len(data[0])

# Animáció inicializálása a diszkrét adatokhoz
line_data, = ax.plot([], [], lw=2, color='black', label='Diszkrét adatok')

# Animáció inicializálása az analitikus megoldáshoz
line_analytical, = ax.plot([], [], lw=2, color='red',alpha = 0.5, label='Analitikus megoldás')

# Ablak beállítása
ax.set_xlim(0, L)
ax.set_ylim(-1, 1)
ax.set_xlabel('Pozíció')
ax.set_ylabel('Kitérés')
ax.set_title('Húr mozgása')
ax.legend()

# Az animáció lépései a diszkrét adatokhoz
def animate(i):
    x = np.linspace(0, L, 100)
    y_data = data[:, i]
    y_analytical_values = y_analytical(x_values, c, L, t_values[i])
    line_data.set_data(x, y_data)
    line_analytical.set_data(x_values, y_analytical_values)
    if i % 10 == 0:
        print(f'Frame {i}/{len(data[0])} processed.')
    return line_data, line_analytical


ani = FuncAnimation(fig, animate, frames=num_steps, interval=100)


ani.save('task2/compare_task2.gif', writer='pillow')

plt.show()
```
