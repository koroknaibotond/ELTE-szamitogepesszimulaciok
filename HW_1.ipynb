```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
def read_txt_file(filename):
    times = []
    positions = []
    velocities = []
    
    with open(filename, 'r') as file:
        for line in file:
            if line.strip():  
                time, position, velocity = map(float, line.split())
                times.append(time)
                positions.append(position)
                velocities.append(velocity)
    
    return np.array(times), np.array(positions), np.array(velocities)



times_euler, positions_euler, velocities_euler = read_txt_file("euler_short.txt")
times_eulercromer, positions_eulercromer, velocities_eulercromer = read_txt_file("eulercromer_short.txt")
```

```python
def Energy_calc(velocities,omega,positions):
    return 0.5 * (velocities**2+omega**2*positions**2)

```

# A harmonikus oszcillátor kitérés-idő diagrammja

```python
# Experiment one - short interval (5 periods)
```

```python
plt.plot(times_euler, positions_euler, label = "Euler")
plt.plot(times_eulercromer,positions_eulercromer, label = "Euler-Cromer")
plt.xlabel("idő [s]")
plt.ylabel("kitérés [cm]")
plt.grid(True)
plt.legend()
plt.xlim(0,10)
plt.title("Kitérés-idő diagramm")
```

```python
# Experiment 2 - long interval (100 periods)
# For a more transparent figure I only plotted the amplitudes in each period
```

```python
times_euler_long, positions_euler_long, velocities_euler_long = read_txt_file("euler_long.txt")
times_eulercromer_long, positions_eulercromer_long, velocities_eulercromer_long = read_txt_file("eulercromer_long.txt")
```

```python
plt.plot(periods,peaks_euler, label = "Euler")
plt.plot(periods,peaks_eulercromer, label = "Euler-Cromer")
plt.grid(True)
plt.xlabel("periódus szorzám")
plt.ylabel("kitérés [cm]")
plt.title("Kitérés-periódus diagramm")
plt.legend()
```

# Futási idő

```python
def read_txt_file_for_time(filename): #needed to rewrite for the new two column format

    steps = []
    times = []

    
    with open(filename, 'r') as file:
        for line in file:
            if line.strip():  # skip empty lines
                step, time = map(float, line.split())
                steps.append(step)
                times.append(time)
               
    
    return np.array(steps), np.array(times)
```

```python
numofsteps, Eulertime = read_txt_file_for_time("eulerTimeConsumption.txt")
numofsteps, EulerCromertime = read_txt_file_for_time("eulerCromerTimeConsumption.txt")
```

```python
plt.plot(numofsteps,Eulertime, label = "Euler")
plt.plot(numofsteps,EulerCromertime, label = "Euler-Cromer")
plt.xlabel("lépsek")
plt.ylabel("idő [ns]")
plt.title(r'$10^6$ lépés esetén')
plt.legend()
plt.grid(True)
```
