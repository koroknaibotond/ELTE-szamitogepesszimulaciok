```python
import matplotlib.pyplot as plt

dt = 0.01  
num_steps = 50000  


with open("task1/T.data", "r") as file:
    temperatures = [float(line.strip()) for line in file]

step_values = range(num_steps)

mean = [np.mean(temperatures[15000:]),np.mean(temperatures[15000:])]
x = [step_values[15000], step_values[-1]]



plt.figure(figsize=(10, 6))
plt.plot(step_values, temperatures, color='black', lw=0.2)    
plt.scatter(step_values[15000], temperatures[15000], color="red", s=50, label = "egyensly kezdete") 
plt.plot(x,mean, color = "red", label ="tlag hmrsklet")
plt.title('Hmrsklet idfejldse', fontsize=16)  
plt.xlabel('Lpsszm', fontsize=14) 
plt.ylabel('Hmrsklet', fontsize=14)  
plt.tick_params(axis='both', labelsize=12)  
plt.grid(True)
plt.legend(fontsize = 12)
plt.show()
```
